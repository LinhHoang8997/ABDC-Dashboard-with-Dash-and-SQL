# An impromptu bubble grid chart

```python
key_order = data_mfg_sum.reset_index().reset_index()[['level_0','mfg_nam']]
```

```python
mean_df = df[['ndc','avg_leak_26']].groupby("ndc").mean().reset_index()
sum_df = df[['ndc','total_leak_26w']].groupby("ndc").sum().reset_index()
mfg_nam_df = df[['ndc','mfg_nam']].drop_duplicates()
ndc_desc_df = df[['ndc','ndc_desc']].drop_duplicates()
combi = pd.merge(mean_df,sum_df,on="ndc",how="left")
combi = pd.merge(combi,ndc_desc_df,on="ndc",how="left")
combi = pd.merge(combi,mfg_nam_df,on="ndc",how="left")
combi = pd.merge(combi,key_order,on="mfg_nam",how="left")
```

```python
combi
```

```python
data_prep = combi.dropna()
```

```python
data_prep = data_prep.rename(columns={'level_0': 'rank'})
```

```python
data_prep = data_prep.reset_index()
```

```python
data_prep['rank'] = data_prep['rank']+1
data_prep
```

## This requires some scaling, or plotly will crash my laptop
We'll use MinMax scaler

The transformation is calculated as:

X_scaled = scale * X + min - X.min(axis=0) * scale
where scale = (max - min) / (X.max(axis=0) - X.min(axis=0))

This transformation is often used as an alternative to zero mean, unit variance scaling.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
new_total_leak = pd.DataFrame(scaler.fit_transform(data_prep[['avg_leak_26','total_leak_26w']]))
```

```python
data_bubble_plot = data_prep.copy()
data_bubble_plot.loc[:,"scaled_total_leak"] = new_total_leak[1]
```

```python
data_bubble_plot.head(2)
```

```python
# bubble plot does not accept zeros. It also does not make sense to include products that do not cost us money from leakage.
data_bubble_plot = data_bubble_plot[data_bubble_plot.total_leak_26w>=0]
data_bubble_plot = data_bubble_plot[data_bubble_plot.avg_leak_26>=0]
```

```python
trace0 = go.Scatter(
    x=data_bubble_plot['rank'],
    y=data_bubble_plot['avg_leak_26'],
    mode='markers',
    marker=dict(
        color=data_bubble_plot['total_leak_26w'],
        size=35,
        opacity=0.9,
        showscale=True
    )
)

data = [trace0]
py.iplot(data, filename='bubblechart-size')
```
